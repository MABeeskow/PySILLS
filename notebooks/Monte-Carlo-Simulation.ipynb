# Monte Carlo simulation

```python
# Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pysills
import csv
```

```python
counter = 0
counter_ppm = 0
counter_xi = 0
counter_intincl = 0
keyword_comp = "COMPOSITIONAL ANALYSIS"
keyword_ppm = "(ppm)"
helper_index = {"COMPOSITIONAL ANALYSIS": [], "(ppm)": []}
keyword_sens = "SENSITIVITY ANALYSIS"
key_xi = "Analytical Sensitivity"
helper_index["SENSITIVITY ANALYSIS"] = []
helper_index["Analytical Sensitivity"] = []
keyword_intens = "INTENSITY ANALYSIS"
key_intincl = "Intensity (Inclusion)"
helper_index["INTENSITY ANALYSIS"] = []
helper_index["Intensity (Inclusion)"] = []

with open("Results_FI.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file)
    for index, line in enumerate(csv_reader):
        if keyword_comp in line:
            counter += 1
            helper_index[keyword_comp].append(index)
        if keyword_ppm in line:
            counter_ppm += 1
            helper_index[keyword_ppm].append(index)
        if keyword_sens in line:
            counter += 1
            helper_index[keyword_sens].append(index)
        if key_xi in line:
            counter_xi += 1
            helper_index[key_xi].append(index)
        if keyword_intens in line:
            counter += 1
            helper_index[keyword_intens].append(index)
        if key_intincl in line:
            counter_intincl += 1
            helper_index[key_intincl].append(index)

#print("The keyword", keyword_comp, "was found", counter, "times.")
#print("Row index:", helper_index[keyword_comp])
#print("")

#print("The keyword", keyword_ppm, "was found", counter_ppm, "times.")
#print("Row index:", helper_index[keyword_ppm])
#print("")

n_files_smpl = helper_index[keyword_ppm][1] - helper_index[keyword_ppm][0] - 4
#print("Number of sample files:", n_files_smpl)
```

```python
index_start = helper_index[keyword_intens][0] + 3
index_end = index_start + n_files_smpl + 2
list_indices = np.arange(index_start, index_end)
print(list_indices)
df_smpl_incl_int = pd.read_csv("Results_FI.csv", skiprows = lambda x: x not in list_indices, delimiter=";")
list_isotopes = list(df_smpl_incl_int.keys())[2:]

df_smpl_incl_int
```

```python
def calculate_concentration(intensity_i, intensity_is, concentration_is, sensitivity_i):
    concentration_i = (intensity_i/intensity_is)*(concentration_is/sensitivity_i)
    return concentration_i
```

```python
list_files = []
for index, filename in df_smpl_incl_int["filename"].items():
    list_files.append(filename)

n_set = 100

for file in list_files:
    print("Current file:", file)
    row_intincl = df_smpl_incl_int.loc[df_smpl_incl_int["filename"] == file].squeeze()[2:]
    row_conc = df_smpl_conc.loc[df_smpl_conc["filename"] == file].squeeze()[2:]
    row_xi = df_smpl_incl_xi.loc[df_smpl_incl_xi["filename"] == file].squeeze()[2:]
    isotope_is = "K39" # row_xi.index[row_xi == 1.0][0]
    norm_row_xi = row_xi/row_xi[isotope_is]
    row_xi = norm_row_xi
    mean_is = row_conc[isotope_is]
    std_is = 0.01*mean_is
    data_is = np.around(np.random.normal(loc=mean_is, scale=std_is, size=n_set), 2)
    #print("IS:", isotope_is, mean_is, std_is)
    #print(row_intincl.tolist())
    #print(row_conc.tolist())
    #print(row_xi.tolist())
    #print(row_xi)
    #print(data_is)
    for isotope in list_isotopes:
        print(isotope)
        intensity_i = row_intincl[isotope]
        intensity_is = row_intincl[isotope_is]
        sensitivity_i = row_xi[isotope]
        concentration_i = round(row_conc[isotope], 3)
        data_intincl_i = np.around(np.random.normal(loc=intensity_i, scale=0.01*intensity_i, size=n_set), 2)
        data_intincl_is = np.around(np.random.normal(loc=intensity_is, scale=0.01*intensity_is, size=n_set), 2)
        data_xi_i = np.around(np.random.normal(loc=sensitivity_i, scale=0.01*sensitivity_i, size=n_set), 2)
        data_conc = np.around(calculate_concentration(data_intincl_i, data_intincl_is, data_is, data_xi_i), 3)
        min_conc = min(data_conc)
        max_conc = max(data_conc)
        std_conc = np.std(data_conc, ddof=1)
        rel_std_conc = std_conc/concentration_i*100
        print("Result from PySILLS:", concentration_i)
        print("Results from Monte Carlo simulation:", "Min:", min_conc, "Max:", max_conc, "Standard deviation:", round(std_conc, 3), "(", round(rel_std_conc, 3), " %)")
```
