#### Packages and modules

```python
import pandas as pd
from pathlib import Path
```

#### Helper functions

```python
def load_icpms_file(path, sep=",", skiprows=3, skipfooter=1, time_name=["Time [Sec]", "time_s"]):
    df = pd.read_csv(path, sep=sep, skiprows=skiprows, skipfooter=skipfooter, engine="python")
    if time_name is not None:
        df = df.rename(columns={time_name[0]: time_name[1]})
    return df
```

#### Data import (from LA-ICP-MS experiments)

```python
base_path = Path("../src/pysills/legacy/lib/demo_files")
df = load_icpms_file(base_path/"demo_ma01.csv")
```
